```python
def statistical_test(df,n,column): 
    # Drop all rows with 0 values
    df = df[df[column] != 0]
    
    # initialize the lists to store the averages
    avg_F = []
    avg_M = []

    for i in range(n):
        # calculate the proportion of each gender in the dataframe
        prop_F = len(df[df['gender'] == 'F']) / len(df)
        prop_M = len(df[df['gender'] == 'M']) / len(df)

        # randomly assign a new gender to each row
        df['gender'] = np.random.choice(['F', 'M'], size=len(df), p=[prop_F, prop_M])

        # calculate and store the average Value for each gender
        avg_F.append(df[df['gender'] == 'F'][column].mean())
        avg_M.append(df[df['gender'] == 'M'][column].mean())
        
    delta = [x - y for x, y in zip(avg_F, avg_M)]
    
    # Plot the distribution of the averages
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    ax1.hist(avg_F, alpha=0.7, density=True, bins=20, label="Simulated avg. female "+column)
    ax1.hist(avg_M, alpha=0.7, density=True, bins=20, label="Simulated avg. male "+column)
    ax1.axvline(df[df['gender'] == 'M'][column].mean(), color='blue', linestyle='dashed', linewidth=2, label = "True avg. female "+column)
    ax1.axvline(df[df['gender'] == 'F'][column].mean(), color='red', linestyle='dashed', linewidth=2, label = "True avg. male "+column)
    ax1.legend( prop={'size': 8})
    ax1.set_title("Distribution of the simulated averages")
    
    ax2.hist(delta, label="F_"+column+" - M_"+column, alpha = 0.7)
    ax2.axvline(df[df['gender'] == 'F'][column].mean() - df[df['gender'] == 'M'][column].mean(), color='red', linestyle='dashed', linewidth=2, label = "True delta_"+column)
    ax2.legend( prop={'size': 8})
    ax2.set_title("Distribution of the simulated delta")
statistical_test(df_Invited,10000,'Age')
```
