```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import matplotlib.cm as cm
import requests
import json
import pandas as pd
import time
import datetime
import os
import sys
import glob
import re
import math
import random
import pickle
import time
import copy
import itertools
import collections
import warnings
warnings.filterwarnings('ignore')
import httplib2
from bs4 import BeautifulSoup, SoupStrainer
```

```python
URL = "https://www.ted.com/talks/quick-list"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
pages = [str(i) for i in range(1,163)]
df = pd.DataFrame(columns = ['Full name',"Year", "Duration"])

for page in pages:
    URL_Name = "https://www.ted.com/talks/quick-list?page=" + page
    page = requests.get(URL_Name)
    soup = BeautifulSoup(page.content, "html.parser")
    content = soup.find_all("div", {"class": "row quick-list__row"})
    Full_name = []
    Year = []
    Length = []

    for row in content:
        speakers = row.find_all("a")
        year = row.find_all('span', {"class":"meta"})
        length = row.find_all("div", {"class": "col-xs-1"})

        # Check result
        # print(speakers[0].text.split(":")[0])
        # print(year[0].text.strip())
        # print(length[1].text.strip())

        # Append data
        Full_name.append(speakers[0].text.split(":")[0])
        Year.append(year[0].text.strip())
        Length.append(length[1].text.strip())
    
    # create a dataframe
    df1 = pd.DataFrame(columns = ['Full name',"Year", "Duration"])
    df1['Full name'] = Full_name
    df1['Year'] = Year
    df1["Duration"] = Length

    time.sleep(2)

    df = df.append(df1, ignore_index=True)
```

```python
df.to_csv("TED_data.csv", index = False)
df.head()
```
